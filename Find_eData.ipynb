```python
#find DNA sequences in the taxonomy_fasta file

#read a fasta file

import sys
import os
from Bio import SeqIO
import pandas as pd
import numpy as np

def convert_fasta_to_df (dir):
    pathToFile = os.path.join(dir)  #windows machine
    allSeqs = []
    for seq_record in SeqIO.parse(pathToFile, """fasta"""):
        allSeqs.append([str(seq_record.seq), str(seq_record.id).split("-")[0]])

    df = pd.DataFrame(np.array(allSeqs))
    df.columns = ['sequence', 'type']
    return df

def convert_fasta_to_df_with_wholeID (dir):
    pathToFile = os.path.join(dir)  #windows machine
    allSeqs = []
    for seq_record in SeqIO.parse(pathToFile, """fasta"""):
        allSeqs.append([str(seq_record.seq), str(seq_record.id)])

    df = pd.DataFrame(np.array(allSeqs))
    df.columns = ['sequence', 'type']
    return df

def preprocess(df_list):
  df_new = []
  for i in df_list:
    A = i[0].count('A')
    T = i[0].count('T')
    G = i[0].count('G')
    C = i[0].count('C')

    if A + T + G + C == len(i[0]):
      seq_li = i[0]
      type_li = i[1]
      df_new.append([seq_li,type_li])
  df = pd.DataFrame (df_new, columns = ['seq', 'type'])
  return df

def filter_weird_alphabet(df_list):
  df_new = []
  for i in df_list:
    A = i[0].count('A')
    T = i[0].count('T')
    G = i[0].count('G')
    C = i[0].count('C')

    if A + T + G + C != len(i[0]):
      seq_li = i[0]
      type_li = i[1]
      df_new.append([seq_li,type_li])
  df = pd.DataFrame (df_new, columns = ['seq', 'type'])
  return df
```
